## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

## 1. 取出今天所有的新聞

```python
import time
todayTime = time.strftime("%Y/%m/%d", time.localtime())
print(todayTime)
```

```python
while True:
    time.sleep(0.5)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    date_tag=soup.find(class_="part_list_2").find_all("h3")
#    print(date_tag[-1].find(class_="date").text[0:10])
    if date_tag[-1].find(class_="date").text[0:10]!=todayTime:
        print("LastTdate:",date_tag[-1].find(class_="date").text[0:10])
        break
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

news=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text[:10] == todayTime:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        news.append({"date":d.find(class_="date").text, 
                     "name":d.find_all('a')[-1].text
                    })        


        
```

```python
import pandas as pd
pd.DataFrame(news)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime, timedelta
TwoHoursAgo = datetime.now()-timedelta(hours=2)

news = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M' ) > TwoHoursAgo:
        
#        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        news.append({"date":d.find(class_="date").text, 
                     "name":d.find_all('a')[-1].text
                    })

```

```python
import pandas as pd
pd.DataFrame(news)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
three_day_time = datetime.now() - timedelta(3)
print("三天前時間：", three_day_time)
date_time_year = three_day_time.strftime("%Y")
date_time_month = three_day_time.strftime("%m")
date_time_day = three_day_time.strftime("%d")
print("三天前時間（年）：", date_time_year)
print("三天前時間（月）：", date_time_month)
print("三天前時間（日）：", date_time_day)


three_days_15_clock = datetime.strptime(three_day_time.strftime("%Y/%m/%d") + ' 15', '%Y/%m/%d %H')
three_days_17_clock = datetime.strptime(three_day_time.strftime("%Y/%m/%d") + ' 17', '%Y/%m/%d %H')
print("三天前下午三點：", three_days_15_clock)
print("三天前下午五點：", three_days_17_clock)
```

```python
# 選擇且點選日期
from selenium.webdriver.support.ui import Select
selY = Select(browser.find_element_by_id("selY"))
selY.select_by_value(date_time_year)
selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(date_time_month)
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(date_time_day)
browser.find_element_by_id('button').click()
```

```python

# 利用 selenium 持續執行下滑的動作，直到取得要求時間內的新聞，表示取得完畢
import time
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(0.5)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    new_date = soup.find(class_="part_list_2").find_all('h3')
    print(new_date[-1].find(class_="date").text)
    if datetime.strptime(new_date[-1].find(class_="date").text, '%Y/%m/%d %H:%M') < three_days_15_clock:
        print('==== STOP ====')
        break
```

```python
news = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    # 比兩小時前早的話就結束
 
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < three_days_17_clock and datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')>three_days_15_clock:
        news.append({
            '時間': d.find(class_="date").text,
            '標題': d.find_all('a')[-1].text
        })
```

```python
import pandas as pd
pd.DataFrame(news)
```
