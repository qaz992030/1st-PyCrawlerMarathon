```python
import requests
from bs4 import BeautifulSoup
```

```python
resp = requests.get('https://www.ettoday.net/news/news-list.htm')
```

```python
soup = BeautifulSoup(resp.text, "html.parser")
```

取出今天所有的發文

```python
titles = soup.find('div' ,'part_list_2').find_all("h3")
titles
```

```python
articles=[]
for title in titles:
    print(title.find(class_="date").text,title.find('em').text ,title.find('a').text)
    articles.append((title.find(class_="date").text,title.find('em').text ,title.find('a').text))
```

果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
```

```python
columns = ['日期', '類別', '標題']
df = pd.DataFrame(articles, columns=columns)
df
```

```python
df.sort_values("類別")
```
