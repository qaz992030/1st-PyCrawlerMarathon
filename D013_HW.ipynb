```python
import requests
from bs4 import BeautifulSoup
```

```python
resp = requests.get("https://www.ptt.cc/bbs/NBA/index.html")
```

```python
soup = BeautifulSoup(resp.text, "html.parser")
soup
```

印出最新文章的「作者」「標題」「時間」

```python
titles = soup.find_all('div', 'r-ent')
titles
```

```python
articles=[]
for title in titles:
    print("{}{}{}" .format(title.find('div','author').text,title.find('div','title').text,title.find('div','date').text))
    articles.append((title.find('div','author').text,title.find('div','title').text,title.find('div','date').text))
```

```python
import pandas as pd
columns = ['作者', '標題', '時間']
df = pd.DataFrame(articles, columns=columns)
df
```
