## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
r = requests.get(" https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text


import json
data = json.loads(response)


len(data)
```

```python
data
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
print("「標題」、「貼文時間」、「留言人數」、「按讚人數」")
for d in data:
    print("{} {} {} {}".format(d['title'],d['createdAt'],d['commentCount'],d['likeCount']))
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''
Comment=[]
Like=[]
for d in data:
    print(d['commentCount'],d['likeCount'])
    Comment.append(d['commentCount'])
    Like.append(d['likeCount'])
print("熱門:平均留言人數{},平均按讚人數{}".format(sum(Comment)/len(Comment),sum(Like)/len(Like)))
  
```
